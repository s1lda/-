1. Взять свою матрицу из 1.10

```python
import numpy as np
from scipy import linalg, stats
M=np.array([[3,-1.2,-8,8],
            [21,-19,0.5,0],
            [7,-4.9,0,-2],
            [1,-2,13,9]])
```

```python
def show(x):
    for i in x:
        for z in i:
            print(f'{round(z,2)}\t',end=' ')
        print()
def showList(x):
    a=''
    for i in x:
        a=a+str(i)+" "
    print(a)
```

2. Получить LU-разложение (LUP-разложение) (факторизацию матрицы)

```python
P,L,U=linalg.lu(M)
print("L:")
show(L)
print("U:")
show(U)
print("P:")
show(P)
```

3. Найти определитель (произведение элементов на главной диагонали L и
U, определителя обратной матрицы к P)

```python
print("det U:")
print(U.diagonal().prod())
print("det L:")
print(L.diagonal().prod())
print("det P:")
print(np.linalg.det(np.linalg.inv(P)))

u1 = U.diagonal().prod()
l1 = U.diagonal().prod()
p1 = np.linalg.det(np.linalg.inv(P))


print(u1*l1*p1)
```

4.Сгенерировать две выборки (вектор целых чисел из 100 элементов) в
одинаковых интервалах с разными распределениями (равномерное и
нормальное)


```python
print("s1:")
sample1 = stats.norm.rvs(size=100)
showList(sample1)
print("s2:")
sample2 = stats.uniform.rvs(size=100)
showList(sample2)
```

5.  Вычислить для каждой из выборок:
a. Среднее;
b. Моду;
c. Медиану;
d. Минимум;
e. Максимум;
f. Стандартное отклонение.

```python
print("mean:")
print(np.mean(sample1), '\t', np.mean(sample2))
print("mode:")
print(stats.mode(sample1, keepdims=True), '\t', stats.mode(sample2, keepdims=True))
print("median:")
print(np.median(sample1), '\t', np.median(sample2))
print("min:")
print(min(sample1), '\t', min(sample2))
print("max:")
print(max(sample1), '\t', max(sample2))
print("std:")
print(np.std(sample1), '\t', np.std(sample2))
```

6.С помощью метода хи-квадрат (scipy.stats.chisquare) вычислить значение
p-value для нулевой гипотезы: “Распределение выборки не
равномерное”.



```python
print(stats.chisquare(sample1))
print(stats.chisquare(sample2))
```
