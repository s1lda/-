Сохраните каждую визуализацию как изображение в формате .png.

```python
n_bins = 20
plt.hist(data['Bedrooms'], bins=n_bins)
plt.savefig('1.png')
```

```python
plt.boxplot(x=data['Bathrooms'])
plt.savefig('2.png')
```

```python

fig, axs = plt.subplots(figsize=(8,8)) # fig будет содержать фигуру, axs - подграфики (в нашем случае один)

count_numbers = data['Balcony'].value_counts() # создаем массив, состоящий из уникальных значений, которые встречаются в этом столбце

axs.pie(count_numbers[1:], autopct='%1.1f%%', startangle=140)
axs.set_title('Карта')
fig.savefig('3.png')
```

```python
#тепловая карта
plt.figure(figsize=(10,15))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt='.2f') # data.corr() - вычисляет корреляцию, annot=True - отображает значения в ячейках, fmt - чтобы после точки отображалось
                                                                 # только 2 цифры.
plt.savefig('4.png')
```

```python
# диаграмма countplot
plt.figure(figsize=(10, 6))
sns.countplot(x='Bedrooms', hue='Lift', data=data)
plt.title('Countplot с группировкой по двум номинативным признакам')
plt.show()
plt.savefig('countplot.png')
```
