17.Отобразите сколько в каком классе пассажиров выжило/не выжило.

```python
df1=data.groupby('pclass')['survived'].count()
df1
```

```python
for i in df1.index:
    print(f"Выжило пассажиров {i}-класса:{data[(data['survived']==1) & (data['pclass']==i)].shape[0]}" )
    print(f"Не выжило пассажиров {i}-класса:{data[(data['survived']==0) & (data['pclass']==i)].shape[0]}" )
```
